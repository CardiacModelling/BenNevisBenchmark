```python
import basin
import numpy as np
import nevis
import random

data = nevis.gb()


def find_labels(h):
    maxima, sn = basin.find_maxima(h)
    label = basin.find_basins(h, sn, maxima)
    return label, maxima

label, maxima = find_labels(data)
```

```python
heights = []

for x, y in maxima:
    heights.append(data[x, y])
```

```python
import matplotlib.pyplot as plt 

plt.hist(heights, log=True, bins=50)
plt.title('Histogram of the heights of local maxima')
plt.xlabel('Height of maxima (m)')
plt.ylabel('Frequency on logscale')
plt.show()
```
